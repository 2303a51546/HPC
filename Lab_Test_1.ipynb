```python
%%bash
# Write the C code to a file
cat <<EOF > program.c
#include <stdio.h>
int main() {
    int n = 10, sum = 0;
    for(int i = 1; i <= n; i++) {
        sum += i;
    }
    printf("Sum = %d\n", sum);
    return 0;
}
EOF
# Compile the C program
gcc program.c -o program

# Run the executable
./program
```

```python
%%bash
# Write the C code to a file
cat <<EOF > program.c
#include <stdio.h>
// Function with more computation (hotspot)
void compute() {
    long long sum = 0;
    for(long long i = 0; i < 100000000; i++) {
        sum += i;
    }
    printf("Compute done\n");
}
// Function with less computation
void process() {
    for(int i = 0; i < 1000000; i++);
    printf("Process done\n");
}
int main() {
    compute();
    process();
    return 0;
}
EOF
# Compile the C program with profiling flags
gcc -pg program.c -o program
# Run the executable to generate gmon.out
./program
```
